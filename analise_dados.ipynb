```python
# Show only Columns
text_cols = df.columns
print(text_cols)
```

```python
# Convert columns from uppercase to lowercase

text_cols = ['Manufacturer', 'Model', 'Vehicle_type', 'Price_in_thousands',
       'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length',
       'Curb_weight', 'Fuel_capacity', 'Fuel_efficiency', 'Latest_Launch',
       'Power_perf_factor', 'mileage']

for text_cols_new in range(len(text_cols)):
    text_cols[text_cols_new]=text_cols[text_cols_new].lower()
print(text_cols)
```

```python
# Show changed columns
df.columns = text_cols
```

```python
# Translating names to pt_br
df.columns = ['fabricante', 'modelo', 'tipo_veiculo', 'preco', 'tam_motor', 'potencia_hp', 'entre_eixos', 'largura', 'comprimento', 'peso', 'cap_combustivel', 'eficiencia', 'ultima_versao', 'fator_de desempenho_de_potência', 'milhagem']
```
